```python
import plotly.plotly as py
import numpy as np
import plotly.graph_objs as go
import pandas as pd
```

```python
df = pd.read_csv('./combinedData.csv')
```

```python
df.head()
```

```python
group = df.groupby(['DAY','COUNTY'])
```

```python
counties = sorted(df['COUNTY'].unique())
counties
```

```python
df2 = group.mean()
df2 = df2.drop(['TOTAL', 'Station number', 'Estimated', 'STATION','RETIRED'],axis=1)
df2.head()
```

```python
days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
df2.loc['Friday']
```

```python
np.array(df2.loc['Friday'].loc['Aitkin'])
```

```python
data = [go.Bar(
            visible = False,
            x=df2.columns,
            y=df2.loc['Sunday'].loc['Aitkin']
    ) for day in days]
data[1]['visible'] = True
layout = go.Layout(
    barmode='group'
)
py.iplot(data, filename='Single Sine Wave',layout=layout)
```

```python
data = []
for day in days:
    for county in counties:
        bar = go.Bar(
            name = county,
            visible = False,
            x=df2.columns,
            y=df2.loc[day].loc[county]
        )
        data.append(bar)
for i in range(41):
    data[i]['visible'] = True
layout = go.Layout(
    barmode='group'
)
py.iplot(data, filename='County Traffic Plot',layout=layout)
```

```python
steps = []
for i in range(7):
    step = dict(
        method = 'restyle',
        args = ['visible', [False] * len(data)],
        label=days[i]
    )
    for k in range(i*41,(i+1)*41):
        step['args'][1][k] = True # Toggle i'th trace to "visible"
    steps.append(step)

sliders = [dict(
    active = 0,
    currentvalue = {"prefix": "Day: "},
    pad = {"t": 50},
    steps = steps
)]

layout = dict(sliders=sliders)

fig = dict(data=data, layout=layout)

py.iplot(fig, filename='County Slider Plot')
```
